# 1. Loading dataset

```python
import pandas as pd
messages = pd.read_csv('sms_spam_db.txt', sep='\t', names=["label", "message"])
```

```python
messages
```

# 2. Data cleaning and preprocessing

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## output/dependent features - label
y = pd.get_dummies(messages['label'])
y = y.iloc[:,1].values ## taking only spam col
```

# 3. Train-test split (BoW)

```python
## train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(corpus, y, test_size=0.20)
```

# 4. Converting text to numerical features (BoW)

```python
## creating BOW model
from sklearn.feature_extraction.text import CountVectorizer
## for binary BOW enabling binary=True
cv = CountVectorizer(max_features=2500, ngram_range=(1,2))
```

```python
len(X_train), len(y_train)
```

```python
## independent features
X_train = cv.fit_transform(X_train).toarray()
X_test = cv.transform(X_test).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, formatter=dict(float=lambda x: "%.3g" % x))
X_train
```

```python
cv.vocabulary_
```

## 5. Training ML model (BoW)

```python
## classification using ml algo
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)
spam_detect_model
```

## 6. Predicting and evaluating (BoW)

```python
y_pred = spam_detect_model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
def preprocess_text(text):
    ps = PorterStemmer()
    text = re.sub('[^a-zA-Z]', ' ', text)  
    text = text.lower().split() 
    text = [ps.stem(word) for word in text if word not in stopwords.words('english')]
    return ' '.join(text) 

input_text = "Congrats! You have cleared the interview. Take the next steps to complete recruitmemt process"
processed_text = preprocess_text(input_text)
input_features = cv.transform([processed_text]).toarray()  
prediction = spam_detect_model.predict(input_features) 
print("Spam" if prediction[0] == 1 else "Ham")
```
